```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib_venn import venn3

macrel = pd.read_csv("../01_macrel/ids_macrel_193.txt", header=None)
macrel = set(macrel[0])
axpep = pd.read_csv("../02_axpep/ids_axpep_193.txt", header=None)
axpep = set(axpep[0])
amp_scanner = pd.read_csv("../03_amp_scanner/ids_amp_scanner_193.txt", header=None)
amp_scanner = set(amp_scanner[0])

venn3([macrel, axpep, amp_scanner], ('Macrel', 'AxPEP', "AMP Scanner v2"))
plt.savefig("venn_diagram.svg")
plt.savefig("venn_diagram.png")
plt.show()
plt.clf()
```
