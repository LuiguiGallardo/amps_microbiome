```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## GRAVY and net charge

```python
# Input amino acid composition
both_datasets = pd.read_csv("./both_groups_gravy_net_charge.tsv", sep="\t")

# Jointplot
sns.set_style("white")
sns.jointplot(data=both_datasets, x="Net charge at pH 7", y="GRAVY index", hue="group")
plt.legend(title="")
plt.xlabel("Net charge")
plt.ylabel("GRAVY index")
plt.savefig("./overexpressed_gravy_net_charge_distribution.svg")
plt.savefig("./overexpressed_gravy_net_charge_distribution.png")
plt.show()
plt.clf()
```
