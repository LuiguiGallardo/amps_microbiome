```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## GRAVY and net charge

```python
from scipy.stats import mannwhitneyu
# Input amino acid composition
both_datasets = pd.read_csv("./both_groups_gravy_net_charge.tsv", sep="\t")

# Jointplot
sns.set_style("white")
sns.jointplot(data=both_datasets, x="Net charge at pH 7", y="GRAVY index", hue="group", palette=["orange", "red"])
plt.legend(title="")
plt.xlabel("Net charge")
plt.ylabel("GRAVY index")

# Draw a line at x=0 and y=0
plt.axhline(y=0, color='lightgray', linestyle='-', zorder=0)
plt.axvline(x=0, color='lightgray', linestyle='-', zorder=0)

plt.savefig("./overexpressed_gravy_net_charge_distribution.svg")
plt.savefig("./overexpressed_gravy_net_charge_distribution.png")
plt.show()
plt.clf()


# Mann-Whitney U test
group_ob_gravy = both_datasets[both_datasets["group"] == "Ob"]["GRAVY index"]
group_oms_gravy = both_datasets[both_datasets["group"] == "OMS"]["GRAVY index"]
stat, p_value = mannwhitneyu(group_ob_gravy, group_oms_gravy)
print("GRAVY index: ", stat, p_value)

group_ob_net = both_datasets[both_datasets["group"] == "Ob"]["Net charge at pH 7"]
group_oms_net = both_datasets[both_datasets["group"] == "OMS"]["Net charge at pH 7"]
stat, p_value = mannwhitneyu(group_ob_net, group_oms_net)
print("Net Charge: ", stat, p_value)

```
