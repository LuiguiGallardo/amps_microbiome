```python
def pie_chart(data, labels, sizes, output_name):
    import matplotlib.pyplot as plt
    import pandas as pd
    import seaborn as sns

    # Set Seaborn style
    sns.set(style="whitegrid")

    data = pd.read_csv(data, sep="\t")

    def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct * total / 100.0))
            return f'{pct:.1f}%\n({val:d})'
        return my_format

    colors = ['lightblue', 'lightgreen', 'plum']
    plt.pie(data[sizes], labels=data[labels], autopct=autopct_format(data[sizes]), colors=colors)
    plt.axis('equal')

    plt.savefig(f'{output_name}.svg')
    plt.savefig(f'{output_name}.pdf')

    plt.show()

if __name__ == '__main__':
    pie_chart('./summary_genomic_context.tsv', "genomic_context", "number_of_amps", "summary_genomic_context")
    pie_chart('./summary_genomic_context_v2.tsv', "genomic_context", "number_of_amps", "summary_genomic_context_v2")
```
