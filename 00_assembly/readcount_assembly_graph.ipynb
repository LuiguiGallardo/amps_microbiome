```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = './readcount_assembly.tsv'

df = pd.read_csv(data, delimiter='\t')
df['Sample'] = df['Sample'].astype(str)

df_melted = pd.melt(df,
                    id_vars=['Sample', 'Group'],
                    var_name='Step',
                    value_name='Value')

# print(df_melted)

ax = sns.barplot(x='Sample',
            y='Value',
            hue='Step',
            data=df_melted)

plt.legend(title='')

plt.ylim(0,16500000)
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.f'))

plt.title('Pretreatment results')
plt.ylabel('Number of paired reads')

plt.savefig("pretreatment_results.svg")
plt.savefig("pretreatment_results.png")
plt.show()
plt.clf()

```
