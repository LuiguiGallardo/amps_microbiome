```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
len_amps = pd.read_csv("./420_transcripts_m2.pep_length_complete.txt")

sns.histplot(len_amps, bins=50)
plt.axvline(x=len_amps.length.mean())
plt.legend("")
plt.title("Predicted AMPs length distribution")
plt.xlabel("Length")
plt.ylabel("Number of AMPs")
plt.tight_layout()
plt.savefig("420_transcripts_m2.pep.svg")
plt.show()
plt.clf()

len_amps_trans = pd.read_csv("./420_transcripts_m2.fna_length_complete.txt")

sns.histplot(len_amps_trans, bins=50)
plt.axvline(x=len_amps_trans.length.mean())
plt.legend("")
plt.title("Predicted transcripts length distribution")
plt.xlabel("Length")
plt.ylabel("Number of Transcripts")
plt.tight_layout()
plt.savefig("420_transcripts_m2.fna.svg")
plt.show()
plt.clf()
```
