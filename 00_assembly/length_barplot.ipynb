```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
len_amps = pd.read_csv("./193_amps.pep_length_complete.txt")

sns.histplot(len_amps, bins=50)
plt.axvline(x=len_amps.length.mean())
plt.legend("")
plt.title("Predicted AMPs length distribution")
plt.xlabel("Length")
plt.ylabel("Number of AMPs")
plt.tight_layout()
plt.savefig("193_amps.svg")
plt.savefig("193_amps.png")
plt.show()
plt.clf()

len_amps_trans = pd.read_csv("./193_amps_trans.fna_length_complete.txt")

sns.histplot(len_amps_trans, bins=50)
plt.axvline(x=len_amps_trans.length.mean())
plt.legend("")
plt.title("Predicted transcripts length distribution")
plt.xlabel("Length")
plt.ylabel("Number of Transcripts")
plt.tight_layout()
plt.savefig("193_amps_trans.svg")
plt.savefig("193_amps_trans.png")
plt.show()
plt.clf()
```
